```python
df = pd.read_csv("mail.csv", encoding='Windows-1252')
```

```python
df
```

```python
unwanted_columns  = ["Unnamed: 2",	"Unnamed: 3",	"Unnamed: 4"]
df.drop(columns = unwanted_columns , inplace= True)
df.head()
```

```python
column_new_name = {"v1": "category" , "v2":"Message"}
df.rename(columns=column_new_name , inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['category'].value_counts()
```

```python
unique_message = df['category'].unique()
message_numbers = {}
i = 0 
for mes in unique_message:
    message_numbers[mes] = i
    i+=1

df['category'] = df['category'].map(message_numbers)
```

```python
df
```

```python
df['category'].value_counts()
```

```python
#lowercasing 
df['Message'] = df['Message'].apply(lambda x:x.lower())
```

```python
#remove puncuation 
import string 
def remove_punc(Message):
    return Message.translate(str.maketrans('','',string.punctuation))
```

```python
df['Message'] = df['Message'].apply(remove_punc)
```

```python
#Remove Numbers 
def remove_numbers(Message):
    new =""
    for i in Message:
        if not i.isdigit():
            new =new+i
    return new 
df['Message'] = df['Message'].apply(remove_numbers)
    
```

```python
#removing urls
import re

def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
```

```python
df['Message'] = df['Message'].apply(remove_urls)
```

```python
#Remove emojies 
def remove_emojis(text):
    return re.sub(r'[^\x00-\x7F]+', '', text)
```

```python
df['Message'] = df['Message'].apply(remove_emojis)
```

```python
#Remove special characters 
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
```

```python
df['Message'] = df['Message'].apply(remove_special_chars)
```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

```python
stopwords = set(stopwords.words("english"))
stopwords 
```

```python
len(stopwords)
```

```python
# lets tokenize 
def remove(Message):
    if isinstance(Message , str): 
        words = word_tokenize(Message)
        cleaned =[]
        for i in words: 
            if i not in stopwords: 
                cleaned.append(i) 
        return ' '.join(cleaned)
    return Message
```

```python
nltk.download("punkt_tab")
```

```python
df['Message'] = df['Message'].apply(remove)
```

```python
#model 
from sklearn.model_selection import train_test_split
```

```python
x = df['Message']
y = df['category']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
#let use bag of words 
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score
```

```python
bag = CountVectorizer()
x_train_bag = bag.fit_transform(x_train)
x_test_bag = bag.transform(x_test)
```

```python
model_1 = MultinomialNB()
model_1.fit(x_train_bag , y_train)
```

```python
predict_bag = model_1.predict(x_test_bag)
```

```python
accuracy_score(y_test , predict_bag)
```

```python
# using TfidfVectorizer
tfidf = TfidfVectorizer() 
x_train_tfidf = tfidf.fit_transform(x_train) 
x_test_tfidf = tfidf.transform(x_test) 
```

```python
model_2 = MultinomialNB()
model_2.fit(x_train_tfidf , y_train)
```

```python
predict_tfidf = model_2.predict(x_test_tfidf)
```

```python
accuracy_score(y_test , predict_tfidf)
```

```python
#logestic regression use 
from sklearn.linear_model import LogisticRegression
```

```python
logestic = LogisticRegression(max_iter=1000)
logestic.fit(x_train_tfidf , y_train)
```

```python
predict_logestic_tfidf = logestic.predict(x_test_tfidf)
```

```python
accuracy_score(y_test , predict_logestic_tfidf)
```

```python
predict_logestic_bag = logestic_bag.predict(x_test_bag)
```

```python
accuracy_score(y_test , predict_logestic_bag)
```
